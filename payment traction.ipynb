```python
from pyspark.sql.functions import *
```

#Payments table

```python
df=spark.sql("select * from samples.wanderbricks.payments")
display(df)
```

```python
df.explain()
```

#Payment summary 
Payment_mentode,status,total_amount,total_tranction

```python
from pyspark.sql.functions import col, sum, count
payments_summary =df.groupBy('payment_method', 'status').agg(sum(col('amount')).alias('total_amount'),count(col('payment_id')).alias('total_transactions')).orderBy('payment_method', 'status','total_amount', ascending=[True, False,False])

display(payments_summary)
```

```python
payments_summary.explain()
```

# Adding Two column 
Total_transaction_amount,Total_transactions _

```python
from pyspark.sql.window import Window
window = Window.partitionBy("payment_method")
payments_summary = payments_summary.withColumn("Total_transaction_amount", sum(col("total_amount")).over(window)).withColumn("Total_transactions _", sum(col("total_transactions")).over(window))
display(payments_summary)
```

```python
payments_summary.explain()
```

# adding percentage columns
percentage_of_amount,Total_transaction_persentage

```python
from pyspark.sql.functions import round

payments_summary = payments_summary.withColumn(
    "percentage_of_amount",
    round(col('total_amount') / col('Total_transaction_amount') * 100, 2)
).withColumn(
    "Total_transaction_persentage",
    round(col('total_transactions') / col('Total_transactions _') * 100, 2)
)
display(payments_summary)
```

```python
payments_summary.explain()
```

```python
payments_summary.explain("cost")
```

```python
payments_summary.selectExpr("spark_partition_id()").distinct().count()
```
