```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
linkes = spark.sql("select * from stockmarket.raw.feedlinks")
newsfeed_link_df = linkes.select("newsfeed", "linkes")
for row in newsfeed_link_df.collect():
    url = row["linkes"]
    try:
        response = requests.get(url)
        display(pd.DataFrame([{
            'newsfeed': row["newsfeed"],
            'link': url,
            'status_code': response.status_code,
            'content_snippet': response.content[:200]
        }]))
    except Exception as e:
        display(pd.DataFrame([{
            'newsfeed': row["newsfeed"],
            'link': url,
            'status_code': 'error',
            'error': str(e)
        }]))
```
