```python
df=spark.sql('''select *,"economictimes" as newsfeed  from stockmarket.sliver.economictimes union select *,"timesofindia" as newsfeed from stockmarket.sliver.timesofindia union  select *,"mint" as newsfeed from stockmarket.sliver.mint 
union  select *,"moneycontrol" as newsfeed from stockmarket.sliver.moneycontrol 
union  select *,"cnbc" as newsfeed from stockmarket.sliver.cnbc 
union select *, "hindubusinessline" as newsfeed from stockmarket.sliver.hindubusinessline''')
```
