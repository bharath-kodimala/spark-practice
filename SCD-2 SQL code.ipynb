# Union all both table as soure table 

```python
%sql
select employ_id as margeID,employ_id,emp_address,'Y' as active_state,current_date as start_date,'2199-01-01' as end_date
 from locala_sample_data.pc.employ_update

 union all

select null as margeID,employ_id,emp_address,'Y' as active_state,current_date as start_date,'2199-01-01' as end_date
 from locala_sample_data.pc.employ_update where exists(select 1 from locala_sample_data.pc.employ_active where employ_id=locala_sample_data.pc.employ_update.employ_id)
 
```

# Using Merge update matched values and inserting unmatched values

```python
%sql
merge into locala_sample_data.pc.employ_active as target_table
using (select employ_id as margeID,employ_id,emp_address,'Y' as active_state,current_date as start_date,'2199-01-01' as end_date
 from locala_sample_data.pc.employ_update

 union all

select null as margeID,employ_id,emp_address,'Y' as active_state,current_date as start_date,'2199-01-01' as end_date
 from locala_sample_data.pc.employ_update where exists(select 1 from locala_sample_data.pc.employ_active where employ_id=locala_sample_data.pc.employ_update.employ_id)
 ) as source_table
on target_table.employ_id=source_table.margeID
when matched then update set target_table.active_state="N",target_table.end_date=current_date

when not matched then insert (employ_id,emp_address,active_state,start_date,end_date) values (source_table.employ_id,source_table.emp_address,source_table.active_state,source_table.start_date,source_table.end_date)


    
```
